What is Label Encoding?

Label Encoding is another way of converting categorical data (like text labels) into numbers, but instead of creating multiple columns (like one-hot encoding), it assigns a unique integer to each category.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```
