```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/Advertising.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
X = data.drop('sales',axis=1)
y = data['sales']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

Testing Error

```python
Testing_error = mean_squared_error(y_test,y_pred)
Testing_error
```

Training Error

```python
y_pred_train = lr.predict(X_train)
Training_error = mean_squared_error(y_train,y_pred_train)
Training_error
```

Ridge L2

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
```

```python
alphavalues = [1e-15,1e-10,1e-8,1e-4,1e-3,1e-2,1,5,10,20,30,100]
solver_values = ['auto','svd']
r = Ridge()
parameters = {'alpha':alphavalues,'solver':solver_values}
ridge_regressor = GridSearchCV(r,parameters)
ridge_regressor.fit(X_train,y_train)
```

```python
ridge_regressor.best_params_
```

```python
ridge_regressor.score(X_train,y_train)
```

```python
ridge = Ridge(alpha=100,solver='auto')
ridge.fit(X_train,y_train)
```

```python
y_pred_ridge = ridge.predict(X_test)
```

```python
y_pred_ridge
```

```python
te = mean_squared_error(y_test,y_pred_ridge)
te
```

```python
ridge.score(X_train,y_train)
```

Lasso **L1**

```python
from sklearn.linear_model import Lasso
lasso = Lasso()

parameters = {'alpha':[1e-15,1e-10,1e-8,1e-4,1e-3,1e-2,1,5,10,20,30,100]}

lasso_regressor = GridSearchCV(lasso,parameters)
lasso_regressor.fit(X_train,y_train)
```

```python
bp= lasso_regressor.best_params_
bp
```

```python
lasso = Lasso(alpha=bp['alpha'])
lasso.fit(X_train,y_train)
```

```python
y_pred_lasso = lasso.predict(X_test)
```

```python
y_pred_lasso
```

```python
lte = mean_squared_error(y_pred_lasso,y_test)
lte
```

```python
lasso.score(X_train,y_train)
```

Elastic Net

```python
from sklearn.linear_model import ElasticNet
elastic = ElasticNet()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-4,1e-3,1e-2,1,5,10,20,30,100]}
elastic_regressor = GridSearchCV(elastic,parameters)
elastic_regressor.fit(X_train,y_train)
```

```python
elastic_regressor.best_params_
```

```python
elastic = ElasticNet(alpha=0.01)
elastic.fit(X_train,y_train)
```

```python
y_pred_elastic = elastic.predict(X_test)
```

```python
y_pred_elastic
```

```python
ete = mean_squared_error(y_test,y_pred_elastic)
ete
```

```python
elastic_score = elastic.score(X_train,y_train)
elastic_score
```

```python
print('Ridge: ', te)
print('Lasso: ', lte)
print('Elastic Net: ', ete)
```
