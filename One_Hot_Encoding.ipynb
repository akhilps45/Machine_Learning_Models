What is One-Hot Encoding?

One-Hot Encoding is a way to convert categorical data (words/labels) into numerical form so that machine learning models can use them.

Since ML models work with numbers, we need to transform text categories like "Red", "Blue", "Green" into numbers â€” but not in a way that gives them unintended order or ranking.
